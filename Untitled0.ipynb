```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
```

```python
# The values here are hypothetical and for demonstration purposes only
data = {
    'User': ['User1', 'User2', 'User3', 'User4', 'User5'],
    'Apache Beam': [5, 4, np.nan, 3, 2],
    'Apache Airflow': [3, np.nan, 3, 2, 4],
    'Google BigQuery': [np.nan, 5, 4, 2, 5],
    'Apache Flink': [4, 3, 5, 4, np.nan],
    'Apache Kafka': [5, np.nan, 4, 4, 3],
}

df = pd.DataFrame(data)
df.set_index('User', inplace=True)
df
```

```python
# Fill not available values with 0
df.fillna(0, inplace=True)

# Compute the cosine similarity matrix
user_similarity = cosine_similarity(df)
user_similarity_df = pd.DataFrame(user_similarity, index=df.index, columns=df.index)
df
```

```python
def similar_users(user, n):
    if user not in df.index:
        return None
    
    # Get the similar users sorted by similarity
    sim_users = user_similarity_df.sort_values(by=user, ascending=False).index[1:]
    
    # Get the top n similar users
    closest_users = sim_users[:n]
    
    return closest_users
```

```python
def recommend(user, n):
    if user not in df.index:
        return None

    closest_users = similar_users(user, n)
    
    # Compute the mean ratings given by the closest users
    mean_ratings = df.loc[closest_users].mean()
    
    # Get the tools that the user hasn't rated yet
    unrated_tools = df.loc[user][df.loc[user].isna()].index
    
    # Get the ratings of the unrated tools by the closest users
    unrated_tools_ratings = mean_ratings[unrated_tools]
    
    # Sort the ratings in descending order and get the top n tools
    recommendations = unrated_tools_ratings.sort_values(ascending=False)[:n]
    
    return recommendations
    print(recommendations)
```
